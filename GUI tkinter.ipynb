```python
import pickle
from tkinter import *
import numpy as np
```

```python
ws = Tk()
ws.title("Crop Prediction System")
ws.geometry('700x500')
ws['bg'] = '#567'

model = pickle.load(open('model.pkl','rb'))

def printValue():
    nitro = nitrogen.get()
    phos = phosphorus.get()
    pota = potassium.get()
    tempe = temperature.get()
    humid = humidity.get()
    ph = phlevel.get()
    rainf = rainfall.get()
    

    feature_list = [nitro, phos, pota, tempe, humid, ph, rainf]
    single_pred = np.array(feature_list).reshape(1, - 1)

    prediction = model.predict(single_pred)
    crop_dict = {1: 'rice', 2: 'maize', 3: 'chickpea', 4: 'kidneybeans', 5: 'pigeonpeas', 6: 'mothbeans', 7: 'mungbean',
                8: 'blackgram', 9: 'lentil', 10: 'pomegranate', 11: 'banana', 12: 'mango',
                13: 'grapes', 14: 'watermelon', 15: 'muskmelon', 16: 'apple', 17: 'orange', 18: 'papaya', 19: 'coconut',
                20: 'cotton', 21: 'jute', 22: 'coffee'}
    if prediction[0] in crop_dict:
        crop = crop_dict[prediction[0]]
        result="{} is a best crop to be cultivated.".format(crop)
    else:
        result="Sorry,we could not determine the best crop to be cultivated with the provided data."

    
    Label(ws, text=f'{result}', pady=20, bg='yellow',foreground='blue',font=0.5).pack()


uname = Label(ws, text = "Enter value of Nitrogen",font=0.5).place(x = 150,y = 20,anchor="center")  
nitrogen = Entry(ws)
nitrogen.pack(pady=10)

uname = Label(ws, text = "Enter value of Phosporus",font=0.5).place(x = 150,y = 60,anchor="center")  
phosphorus = Entry(ws)
phosphorus.pack(pady=10)

uname = Label(ws, text = "Enter value of Potassium",font=0.5).place(x = 150,y = 100,anchor="center")  
potassium = Entry(ws)
potassium.pack(pady=10)

uname = Label(ws, text = "Enter value of Temperature",font=0.5).place(x = 150,y = 140,anchor="center")  
temperature = Entry(ws)
temperature.pack(pady=10)

uname = Label(ws, text = "Enter value of humidity",font=0.5).place(x = 150,y = 180,anchor="center")  
humidity = Entry(ws)
humidity.pack(pady=10)

uname = Label(ws, text = "Enter value of PH",font=0.5).place(x = 150,y = 220,anchor="center")  
phlevel = Entry(ws)
phlevel.pack(pady=10)

uname = Label(ws, text = "Enter value of Rainfall",font=0.5).place(x = 150,y = 260,anchor="center")  
rainfall = Entry(ws)
rainfall.pack(pady=10)

Button(
    ws,
    text="Recommended Crop",
    font=0.5,
    padx=10,
    pady=10,
    foreground='white',
    bg='seagreen',
    command=printValue
    ).pack(pady=10)

ws.mainloop()
```
